```python
from anomalib.data import MVTec
from anomalib.models import Patchcore, EfficientAd
from anomalib.engine import Engine
```

```python
def train_and_evaluate(model_class, categories):
    results = {}
    for category in categories:
        print(f"Training and evaluating {model_class.__name__} on {category}")
        datamodule = MVTec(category=category, train_batch_size=1)
        model = model_class()
        engine = Engine(max_epochs=5)
        engine.fit(datamodule=datamodule, model=model, ckpt_path="results/EfficientAd/MVTec/grid/v0/weights/model.ckpt")
        scores = engine.test(datamodule=datamodule, model=model)
        print(scores)
        results[category] = scores
    
    return results, model

categories = ["tile", "leather", "grid"]
```

```python
# Patchcore
patchcore_results, patchcore_model = train_and_evaluate(Patchcore, categories)
```

```python
# EfficientAD
efficientAd_results, efficientAd_model = train_and_evaluate(EfficientAd, categories)
```

```python
print('----------------------------------')

print("PatchCore Results")
for key, value in patchcore_results.items():
    print(key, " : ", value[0]['image_AUROC'])

print('----------------------------------')

print("EfficientAD Results")
for key, value in efficientAd_results.items():
    print(key, " : ", value[0]['image_AUROC'])

print('----------------------------------')
```

#### Extracting the feature for the query image

```python
import numpy as np
import torch.nn.functional as F
from sklearn.decomposition import PCA
from qdrant_client import QdrantClient
from qdrant_client.models import PointStruct
from qdrant_client.http import models as qdrant_models

collection_name = 'anomaly_features'

pca = PCA(n_components=50)

def extract_features(model, image, image_path):
    print(image_path)
    image_id = image_path.split('/')[-1].split('.')[0]
    print(image.shape)
    embeddings = model(image)
    print(embeddings.shape)
    reduced_embeddings = pca.fit_transform(embeddings)
    feature = reduced_embeddings.reshape(-1)
    return feature, image_id

query_feature, image_id = extract_features(models["patch_core"], image=query_image, image_path=query_image_path)
print(query_feature.shape)
```

```python
def similarity_search(image_features, top_k=5):
    result = client.search(
        collection_name=collection_name,
        query_vector=image_features,  # Feature vector of the input image
        limit=top_k  # Number of similar images to return
    )
    return result

# Example usage: Assuming `new_image_features` is the feature vector of a new image
results = similarity_search(query_feature)
print(results)
```
